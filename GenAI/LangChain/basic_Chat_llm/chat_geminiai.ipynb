```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.3
)

response = llm.invoke("What is Docker?")

print(response.content)
```

```python
import google.generativeai as genai

genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))

for m in genai.list_models():
    print(m.name)
```

```python
# Language Translation
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
messages = [
    SystemMessage(content="You are a helpful assistant that translates English to Hindi."),
    HumanMessage(content="What is your name?"),]
llm.invoke(messages)
```

```python
messages = [
    SystemMessage(content="You are a helpful assistant that translates Hindi to English."),
    HumanMessage(content="तुम्हारा नाम क्या है?"),]
llm.invoke(messages)  
```

```python
llm.invoke("What is the capital of India?")
llm.invoke([{"role": "system", "content": "You are a helpful assistant that translates English to Hindi."},
            {"role": "user", "content": "What is your name?"}]) 
```

```python
# Prompt Template
from langchain_core.prompts import ChatPromptTemplate
system_template = "Translate the following from English into {language}"
prompt_template = ChatPromptTemplate.from_messages(
    [("system", system_template), ("user", "{text}")]
)
prompt = prompt_template.invoke({"language": "odia", "text": "how are you"})
prompt
```

```python
prompt.to_messages()
```

```python
response = llm.invoke(prompt)
print(response.content)
```
